```python
import pandas as pd
```

```python
predicted_bestsellers = pd.read_csv('data/bestsellers_predicted.csv')
```

```python
nan_titles = predicted_bestsellers['Tytuł oryginału'].isnull()
```

```python
predicted_bestsellers.loc[nan_titles, 'Tytuł oryginału'] = predicted_bestsellers.loc[nan_titles, 'url'].str.split('/').str[-1].str.split('-').str.join(' ').str.title()
```

```python
predicted_bestsellers.columns
```

```python
predicted_bestsellers.loc[predicted_bestsellers['number_of_people_has'] < 2000, ['Tytuł oryginału', 'author', 'Liczba stron', 'description', 'number_of_people_has']].sort_values('number_of_people_has', ascending=True)
```
